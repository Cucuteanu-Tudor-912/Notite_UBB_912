```python
import torch
import torch.nn as nn
from collections import OrderedDict
```

```python
# your code here
#model1 = nn.Sequential(OrderedDict([
#    ('hidden', nn.
#]))
model1 = nn.Sequential(OrderedDict([
    ('hidden', nn.Linear(2, 2)),
    ('activation', nn.ReLU()),
    ('output', nn.Linear(2, 2))
]))
```

```python
print(model1)
```

```python
# your code here
data_in = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
print(data_in)
```

```python
data_target = torch.tensor([[0,0], [0,1], [0,1], [1,0]], dtype=torch.float32)
print(data_target)
```

```python
# your code here
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model1.parameters(), lr=0.1)
```

```python
# your code here
# Train the model
for epoch in range(1000):
    # Forward pass: Compute predicted y by passing x to the model
    y_pred = model1(data_in)

    # Compute and print loss
    loss = criterion(y_pred, data_target)
    print(f'Epoch {epoch + 1}/1000 | Loss: {loss.item()}')

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    # if loss.item() < 1e-3:
    if loss.item() < 1e-3:
        break
```

```python
# your code here
# visualize the results
print(model1(data_in))
```

```python
# calculate the accuracy 
accuracy = 0
for i in range(len(data_target)):
    difference = (model1(data_in)[i] - data_target[i]).abs()
    if torch.all(difference < 0.5):
        accuracy += 1

# print it as a percentage
print(f'Accuracy: {accuracy/len(data_target) * 100}%')
```

```python
import torch
import torch.nn as nn
from collections import OrderedDict

# Define the dataset
X_train = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)  # Training data
y_train = torch.tensor([[0, 0], [0, 1], [0, 1], [1, 0]], dtype=torch.float32)  # Labels
input_size = 2
output_size = 1
# Define architectures for three different models
architectures = [
    OrderedDict([
        ('hidden1', nn.Linear(2, 2)),
        ('activation1', nn.ReLU())
        # ('output', nn.Linear(2, 1))
    ]),
    OrderedDict([
        ('hidden1', nn.Linear(2, 4)),
        ('activation1', nn.ReLU()),
        ('output', nn.Linear(4, 2)),
        ('activation2', nn.ReLU())
        # ('output', nn.Linear(2, 1))
    ]),
    OrderedDict([
        ('hidden1', nn.Linear(2, 8)),
        ('activation1', nn.ReLU()),
        ('hidden2', nn.Linear(8, 4)),
        ('activation2', nn.ReLU()),
        ('output', nn.Linear(4, 2)),
        ('activation3', nn.ReLU())
        # ('output', nn.Linear(2, 1))
    ])
]


# Training configurations
learning_rates = [0.1, 0.1, 0.1]
num_epochs = 1000

# Train and evaluate each model
best_accuracy = 0
best_model = None
best_epoch = 0
best_idx = 0
for idx, (architecture, learning_rate) in enumerate(zip(architectures, learning_rates)):
    # Define the model
    model = nn.Sequential(architecture)
    criterion = nn.MSELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
    cur_epoch = 0
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        y_pred = model(X_train)
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        
        if loss.item() < 1e-3:
            cur_epoch = epoch
            break
        if (epoch + 1) % 100 == 0:
            print(f'Epoch {epoch + 1}/{num_epochs} | Loss: {loss.item()}')
    
    if cur_epoch == 0:
        cur_epoch = num_epochs
    predicted = torch.clamp(model(X_train).round(), 0, 1)
    accuracy = (predicted == y_train).all(dim=1).float().mean().item() * 100
    print(f'Model {idx+1} Accuracy: {accuracy}%')
    
    if accuracy >= best_accuracy:
        ok = True
        if accuracy == best_accuracy and cur_epoch >= best_epoch:
            ok = False
        if ok:
            best_accuracy = accuracy
            best_model = model
            best_epoch = cur_epoch
            best_idx = idx

# Display the weights for each layer of the best model
print("\nWeights for the best model:")
for name, param in best_model.named_parameters():
    print(name, param.data)

print("\nBest model state dict:")
print(best_model.state_dict())
print(f'Best model: {best_idx + 1} with accuracy {best_accuracy}% and trained for {best_epoch} epochs')
```

```python
import torch
import torch.nn as nn
from collections import OrderedDict

# Define the dataset
X_train = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)  # Training data
y_train = torch.tensor([[0, 0], [0, 1], [0, 1], [1, 0]], dtype=torch.float32)  # Labels
input_size = 2
output_size = 1
# Define architectures for three different models
architectures = [
    OrderedDict([
        ('hidden1', nn.Linear(2, 3)),
        ('activation1', nn.ReLU()),
        # ('hidden2', nn.Linear(3, 2)),
        ('output', nn.Linear(3, 2)),
        ('activation2', nn.ReLU())
        # ('activation2', nn.ReLU()),
        # ('output', nn.Linear(2, 1))
        
    ]),
    OrderedDict([
        ('hidden1', nn.Linear(2, 3)),
        ('activation1', nn.ReLU()),
        # ('hidden2', nn.Linear(3, 2)),
        ('output', nn.Linear(3, 2)),
        ('activation2', nn.Tanh())
        # ('activation2', nn.Tanh()),
        # ('output', nn.Linear(2, 1))
    ]),
    OrderedDict([
        ('hidden1', nn.Linear(2, 3)),
        ('activation1', nn.ReLU()),
        # ('hidden2', nn.Linear(3, 2)),
        ('output', nn.Linear(3, 2)),
        ('activation2', nn.Sigmoid())
        # ('activation2', nn.Sigmoid()),
        # ('output', nn.Linear(2, 1))
    ])
]

# Training configurations
learning_rates = [0.1, 0.1, 0.1]
num_epochs = 10000

# Train and evaluate each model
best_accuracy = 0
best_model = None
best_epoch = 0
best_idx = 0
for idx, (architecture, learning_rate) in enumerate(zip(architectures, learning_rates)):
    # Define the model
    model = nn.Sequential(architecture)
    criterion = nn.MSELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
    cur_epoch = 0
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        y_pred = model(X_train)
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        
        if loss.item() < 1e-3:
            cur_epoch = epoch
            break
        if (epoch + 1) % 100 == 0:
            print(f'Epoch {epoch + 1}/{num_epochs} | Loss: {loss.item()}')
    
    if cur_epoch == 0:
        cur_epoch = num_epochs
    predicted = torch.clamp(model(X_train).round(), 0, 1)
    accuracy = (predicted == y_train).all(dim=1).float().mean().item() * 100
    print(f'Model {idx+1} Accuracy: {accuracy}%')
    
    if accuracy >= best_accuracy:
        ok = True
        if accuracy == best_accuracy and cur_epoch >= best_epoch:
            ok = False
        if ok:
            best_accuracy = accuracy
            best_model = model
            best_epoch = cur_epoch
            best_idx = idx

# Display the weights for each layer of the best model
print("\nWeights for the best model:")
for name, param in best_model.named_parameters():
    print(name, param.data)

print("\nBest model state dict:")
print(best_model.state_dict())
print(f'Best model: {best_idx + 1} with accuracy {best_accuracy}% and trained for {best_epoch} epochs')
```
